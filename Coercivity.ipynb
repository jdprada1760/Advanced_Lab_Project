```python
# Fit function for Descending part of Hysteresis (data1)
def hysDesc(x,a,b,c,d,e):
    return -(a*arctan(b*x+c)-d) + e*x
# Fit function for Ascending part of Hysteresis (data2)
def hysAsc(x,a,b,c,d,e):
    return a*arctan(b*x+c)+d + e*x
# Optimize fit functions
params1, errors1 = optimize.curve_fit(hysDesc,data1[0],data1[1],  maxfev = 100000)
params2, errors2 = optimize.curve_fit(hysAsc,data2[0],data2[1], maxfev = 100000)
print params1,params2
```

```python
# Fit function for Descending part of Hysteresis (data1)
def lin(x,a,b):
    return a+ b*x
# Fit function for Ascending part of Hysteresis (data2)
def hysAsc(x,a,b,c,d,e):
    return a*arctan(b*x+c)+d + e*x
# Defines chi^2 of lin functtion
def chiLin(x,y,a,b):
    yf = lin(x,a,b)
    tmp = (yf-y)**2
    return mean(tmp)
# Defines chi^2 of hysAsc functtion
def chiHys(x,y,a,b,c,d,e):
    yf = hysAsc(x,a,b,c,d,e)
    tmp = (yf-y)**2
    return mean(tmp)
    
# Fits hysteresis functions(ASC DESC) for a given filename and finds coercivity
# if booli= true => plots fit
cut = 80
def coercivity(filename,booli):
    filename = dirn+filename
    # Loads M vs H data (omits first 12 rows)
    # First data row is usually  biased by change of scale
    data = np.loadtxt(filename, skiprows = 13)
    # Separates data (Descending + Ascending)
    # Rotates descending one to match the ascending
    newdat = [ i for i in -data[:cut]] 
    for j in data[cut:]:
        newdat.append(j)
    newdat = np.array(newdat).T
    #print newdat
    #plt.ylim(-0.0005,0.0005)
    # Optimize fit functions
    params1, errors1 = optimize.curve_fit(hysAsc,newdat[0],newdat[1],  maxfev = 1000000)
    params11, errors11 = optimize.curve_fit(lin,newdat[0],newdat[1],  maxfev = 1000000)
    #print params1,params2
    if (booli):
        fig = plt.figure(figsize = (7,5))
        ax = fig.add_subplot(111)
        ax.plot(newdat[0],newdat[1],'o')
        ax.plot([-800,800],[0,0])
        ax.set_title(filename.split("#")[0].split("Ni")[1],fontsize= 30)
        Bx = linspace(min(newdat[0])-20,max(newdat[0])+20,200)
        if( chiLin(newdat[0],newdat[1],params11[0],params11[1]) >= chiHys(newdat[0],newdat[1],params1[0],params1[1],params1[2],params1[3],params1[4]) ):
            ax.plot(Bx, hysAsc(Bx,params1[0],params1[1],params1[2],params1[3],params1[4]))
        else:
            ax.plot(Bx, lin(Bx,params11[0],params11[1]))
    def fun(x):
        return hysAsc(x,params1[0],params1[1],params1[2],params1[3],params1[4])
    # Finds coercivity 1 and 2
    sol1 = optimize.root(fun, x0 = 0.).x
    if( chiLin(newdat[0],newdat[1],params11[0],params11[1]) <= chiHys(newdat[0],newdat[1],params1[0],params1[1],params1[2],params1[3],params1[4]) ):
        sol1 = -params11[0]/params11[1]
    #print filename
    #print sol1
    return sol1
```

```python
cut = 80
coercivity(dirl[0],True)
```

```python
# Loads dir filenames list
dirn = "./Data/Perp_RoomTemp/"
dirl = os.listdir(dirn)
#print dirl
#print dirl
# Obtains angles from filenames
angles = np.array([float(a.split("#")[0].split("Ni")[1])  for a in dirl ])
# Array of coercivities
coercvty = np.array([coercivity(fille,True) for fille in dirl])
```

```python
#print coercvty
fig = plt.figure(figsize = (15,10))
ax = fig.add_subplot(111)
ax.plot(angles,coercvty, 'o')
ax.set_ylabel(r"$Coercivity(G)$",fontsize = 30)
ax.set_xlabel(r"$Angle(^o)$",fontsize = 30)
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=20)
ax.set_xlim(0,190)
ax.set_title(r"$First Measurement$",fontsize = 40)
```
